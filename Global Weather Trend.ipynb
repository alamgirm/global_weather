# Global Weather Trend
In this project, local and global temperature data are analyzed and compared to the temperature trends for a city to overall global temperature trends.
![Earth](earth.png "Earth")

```python
#load all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
global_data = pd.read_csv('global_data.csv')
city_data = pd.read_csv('city_data.csv')
city_list = pd.read_csv('city_list.csv')
```

```python
# select data for Toronto
toronto_data = city_data[city_data['city']=='Toronto'].reset_index()
toronto_data.drop(['city', 'country', 'index'], axis=1, inplace=True)
toronto_data.head()
```

```python
# take a 7-year rolling mean
td = toronto_data.rolling(min_periods=7,window=7,center=False).mean()

fig, axe = plt.subplots(1,1, figsize=(15,5))
years_t = td.year.tolist()
tdata_t = td.avg_temp.tolist()
axe.plot(years_t, tdata_t, label='Toronto')
```

```python
gd = global_data.rolling(min_periods=7,window=7,center=False).mean()
```

```python
plt.figure(1, figsize=(15,5))
plt.plot(years_t, tdata_t, label='Toronto')
years_g = gd.year.tolist()
tdata_g = gd.avg_temp.tolist()
plt.plot(years_g, tdata_g, label='Global')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Avg Temp')
plt.title('7-Year Moving Average of Mean Temperature')
```
